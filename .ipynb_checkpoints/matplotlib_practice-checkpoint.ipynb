```python
#import matplotlib.pyplot and shorten the call for ease of use
#note: pyplot provides MATLAB-ish interface
import matplotlib.pyplot as plt
```

```python
#add small portion of ride sharing data for graph

#create set for x-axis - they will be months
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#create set for y-axis - total fare in US dollars per month
#note: float values
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#line charts
#plot the values in the x_axis and y_axis sets
plt.plot(x_axis,y_axis, marker="*", color="blue", linewidth=2, label='Boston')

#add labels and chart title
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare per Month (2014)')

#set y limit between 0 and 45
plt.ylim(0,45)

#add a legend
plt.legend()

#add a grid
plt.grid()
```

```python
#chart using object oriented interface versus previous MATLAB type
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#V2 AX.PLT create the plot with ax.plt()
#note: this version breaking fig and ax apart versus previous
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
#V3 AX.PLT create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis,marker="D", color="green", linewidth=2, label='Boston')

#add labels and chart title
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare per Month (2014)')

#set y limit between 0 and 45
ax.set_ylim(0,45)

#add a legend
ax.legend()

#add a grid
ax.grid()
```

```python
#bar charts
plt.bar(x_axis,y_axis, color="green", label='Boston')

#add labels and chart title
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare per Month (2014)')

#set y limit between 0 and 45
plt.ylim(0,45)

#add a legend
plt.legend()
```

```python
#horizontal bar chart
plt.barh(x_axis,y_axis)
```

```python
#horizontal bar chart
#invert the values on the y axis
plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.gca().invert_yaxis()
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare per Month (2014)')
plt.legend()
```

```python
#bar charts using object oriented approach
#ax.plt() bar
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
#bar charts using object oriented approach
#ax.plt() bar
#horizontal bar
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color="cyan",label="Chicago")

#add labels and chart title
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare per Month (2014)')
#add legend
ax.legend()

#invert y axis
ax.invert_yaxis()
```

```python
#scatter plots
#adding o to plt.plot method
plt.plot(x_axis,y_axis,'o')
```

```python
#scatter plot using plt.scatter
plt.scatter(y_axis,x_axis,color="red",label="Chicago")
plt.gca().invert_yaxis()
plt.xlim(0,45)
plt.ylabel('Date')
plt.xlabel('Fare ($)')
plt.title('PyBer Fare per Month (2014)')
plt.legend()
```

```python
#scatter plot using plt.scatter
#make the dots bigger by creating a new list that is y_axis 3x
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
    
#markers graduated based on fare value
plt.scatter(x_axis,y_axis, s=y_axis_larger)
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare per Month (2014)')
```

```python
#scatter plot using plt.scatter
#make the dots bigger by using list comprehension
#markers graduated based on fare value
plt.scatter(x_axis,y_axis, s=[i*3 for i in y_axis])
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare per Month (2014)')
```

```python
#scatter plot using object-oriented approach
fig,ax=plt.subplots()
ax.scatter(y_axis,x_axis,s=[i*5 for i in y_axis],color="skyblue",edgecolor="black",linewidths=2,label="Boston")

#legend, limit and chart title
ax.legend()
ax.set_xlim(0,45)
ax.set_title('PyBer Fare per Month (2014)')

#invert y axis
ax.invert_yaxis()
```

```python
#pie charts
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#make the pie chart bigger to account for percentage labels crowding
plt.subplots(figsize=(8,8))

#pie chart with exploding wedge for the seventh value (july)
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)


#assign colors for the months
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#adding exploding wedge and percentages to the pies
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
#pie charts using object oriented approach
#standard obj oriented beginning with figsize of 8x8 added
fig, ax = plt.subplots(figsize=(8,8))

#explode values for top two months - march and july - indeces 2 and 6
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#assign colors for the months
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#build pie chart
ax.pie(y_axis,
       labels = x_axis,
       explode = explode_values,
       autopct = '%.1f%%',
       colors = colors,
       shadow = True,
       counterclock = False, 
       startangle=90)
```
